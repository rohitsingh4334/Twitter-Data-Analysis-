```python
import json


Fresh_tweets = 'twitterDB2.csv'
tweets_data = []
with open(Fresh_tweets,'r') as tweets_file:
    for line in tweets_file:
        try:
            tweet= json.loads(line)
            tweets_data.append(tweet)
        except:
            continue
print len(tweets_data)
```

```python
#import Data_List as l
import pandas as pd

import matplotlib.pyplot as plt

data=tweets_data
tweets = pd.DataFrame()
tweets['text'] = map(lambda tweet : tweet['text'],data)
tweets['lang'] = map(lambda tweet: tweet['lang'],data)
tweets['country'] = map(lambda tweet: tweet['place']['country'] if tweet['place'] != None else None,data)
csv_data = {'Text':tweets['text'],'language':tweets['lang'],'Country':tweets['country']}
df=pd.DataFrame(csv_data,columns=['Text','language','Country'])
print "Show table!!!!!"
print df.head(10)
tweets_by_lang = tweets['lang'].value_counts()
```

```python
###Graph of Top languages with counts bar graph###
df=pd.DataFrame(csv_data,columns=['Text','language'])
print"\ntable to show deatils with languaes!!!!"
print df.head(10)
print "\n\t tweets by top 5 language\n"
print tweets_by_lang.head(5)
fig,ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Languages', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 languages', fontsize=15, fontweight='bold')
tweets_by_lang[:5].plot(ax=ax, kind='bar', color='red')
plt.show()
```

```python
import re
def search_word(word,text):
    word=word.lower()
    text=text.lower()
    match=re.search(word,text)
    if match: 
        print text
        return True
    else:   
        return False

#add data
tweets['java'] = tweets['text'].apply(lambda tweet: search_word('java', tweet))
tweets['python'] = tweets['text'].apply(lambda tweet: search_word('python', tweet))
tweets['ruby'] = tweets['text'].apply(lambda tweet: search_word('ruby', tweet))
```

```python
program_langs=['java','python','ruby']
tweets_program_langs=[tweets['java'].value_counts()[True],tweets['python'].value_counts()[True],tweets['ruby'].value_counts()[True]]

print "Java",tweets['java'].value_counts()[True]
print "Python",tweets['python'].value_counts()[True]
print "Ruby",tweets['ruby'].value_counts()[True]

#print tweets_program_langs

x_pos = list(range(len(program_langs)))
width = 0.8
fig, ax = plt.subplots()
plt.bar(x_pos, tweets_program_langs, width, alpha=1, color='g')

# Setting axis labels and ticks
ax.set_ylabel('Number of tweets', fontsize=15)
ax.set_title('Ranking: java vs. python and ruby', fontsize=10, fontweight='bold')
ax.set_xticks([p + 0.4 * width for p in x_pos])
ax.set_xticklabels(program_langs)
plt.grid()
print "\n\t  Working graph for Ranking: java vs. python vs ruby !!"
plt.show()

###Graph of Top languages with counts pie chart###
# Data to plot
labels = program_langs
sizes = tweets_program_langs
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.1)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
tweets['programming'] = tweets['text'].apply(lambda tweet: search_word('programming', tweet))
tweets['job'] = tweets['text'].apply(lambda tweet: search_word('job', tweet))
tweets['developer']=tweets['text'].apply(lambda tweet: search_word('developer', tweet))
tweets['learning']=tweets['text'].apply(lambda tweet: search_word('learning', tweet))

#print tweets['developer']
```
